```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
# Set plot style
plt.style.use('ggplot')
```

```python
df = pd.read_csv("Final_cleaned_data.csv")
```

```python
df.head()
```

```python
# 1. Distribution of Book Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Book_Price'], bins=30, kde=True)
plt.title('Distribution of Book Prices')
plt.xlabel('Book Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Popularity Score vs. Book Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Book_Price'], y=df['Popularity_Score'], alpha=0.6)
plt.title('Popularity Score vs. Book Price')
plt.xlabel('Book Price')
plt.ylabel('Popularity Score')
plt.show()
```

```python
# 3. Number of Books Published Per Year
plt.figure(figsize=(12, 6))
sns.histplot(df['First_Publish_Year'], bins=30, kde=True)
plt.title('Number of Books Published Per Year')
plt.xlabel('First Publish Year')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Bestseller Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df['Bestseller'])
plt.title('Bestseller Distribution')
plt.xlabel('Bestseller (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Book Length Category Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df['Book_Length_Category'], order=df['Book_Length_Category'].value_counts().index)
plt.title('Book Length Category Distribution')
plt.xlabel('Book Length Category')
plt.ylabel('Count')
plt.show()
```

```python
# 6. Total Reviews vs. Popularity Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Total_Reviews'], y=df['Popularity_Score'], alpha=0.6)
plt.title('Total Reviews vs. Popularity Score')
plt.xlabel('Total Reviews')
plt.ylabel('Popularity Score')
plt.show()
```

```python
# 7. Top 10 Publishers by Book Count
plt.figure(figsize=(12, 6))
top_publishers = df['Publisher'].value_counts().nlargest(10)
sns.barplot(x=top_publishers.index, y=top_publishers.values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Publishers by Book Count')
plt.xlabel('Publisher')
plt.ylabel('Number of Books')
plt.show()
```

```python
# 8. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Book Features')
plt.show()
```

```python
# 9. Edition Count Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Edition_Count'], bins=30, kde=True)
plt.title('Distribution of Edition Counts')
plt.xlabel('Edition Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# 10. Popularity Score by Book Length Category
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Book_Length_Category'], y=df['Popularity_Score'])
plt.title('Popularity Score by Book Length Category')
plt.xlabel('Book Length Category')
plt.ylabel('Popularity Score')
plt.show()
```
