```python
import requests
import pandas as pd

# Load the BX-Books dataset
bx_books_path = "/content/BX-Books.csv"
df_bx_books = pd.read_csv('BX-Books.csv', sep='\";\"', names=['ISBN', 'Book-Title' ,'Book-Author','Year-Of-Publication', 'Publisher', 'Image-Url-S', 'Image-Url-M', 'Image-Url-L'], encoding='latin-1', skiprows=1)

# Extract book titles
book_titles = df_bx_books["Book-Title"].dropna().unique()

# Function to fetch book details from Open Library API
def get_book_data(title):
    url = f"https://openlibrary.org/search.json?title={title.replace(' ', '+')}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if "docs" in data and len(data["docs"]) > 0:
            book = data["docs"][0]  # Get first result
            return {
                "Title": title,
                "Author": ", ".join(book.get("author_name", ["Unknown"])),
                "Subjects": ", ".join(book.get("subject", ["None"])),
                "First Publish Year": book.get("first_publish_year", "N/A"),
                "Edition Count": book.get("edition_count"),
                "ISBN": book.get("isbn", ["N/A"])[0] if "isbn" in book else "N/A"
            }
    return {"Title": title, "Author": "Not Found", "Subjects": "Not Found", "First Publish Year": "N/A", "Edition Count": "N/A", "ISBN": "N/A"}

# Fetch data for a subset of book titles (adjustable)
book_data_list = [get_book_data(title) for title in book_titles[9000:10000]]  # Adjust number of books

# Save extracted data to CSV
df_books_api = pd.DataFrame(book_data_list)
output_file = "/content/extracted_books_data9.csv"
df_books_api.to_csv(output_file, index=False)

print(f"Book details extracted and saved to {output_file}")
```
